### Find the number of titles that have changed in the United States Code since its last release point: url = 'http://uscode.house.gov/download/download.shtml'

```python
url = 'http://uscode.house.gov/download/download.shtml'
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    target_element = soup.find_all('div', class_='usctitlechanged')
    print("Number of titles changed:", len(target_element))

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

### Display the top 10 languages by number of native speakers stored in a pandas dataframe: url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'

```python
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
response = requests.get(url)
row_count = 0
language = []
numbers = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.find('table', class_='wikitable')
    
    for row in table.find_all('tr')[1:]:
        columns = row.find_all(['td','th'])
        language.append(columns[0].text.strip())
        numbers.append(columns[1].text.strip())
        
        row_count+=1
        if row_count >= 10:
            break
    
    languages = pd.DataFrame({'Language':language, 'Native Speakers':numbers})
    print(languages)

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

```

### A list with the different kind of datasets available in data.gov.uk: url = 'https://data.gov.uk/'

```python
##main-content > div:nth-child(3) > div > ul

url = 'https://www.data.gov.uk/'
response = requests.get(url)
datasets = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.select("#main-content > div:nth-child(3) > div > ul > li > h3 > a")
    #print(table)
    
    for title in table:
        #print(title)
        name = title.get_text()
        link = title['href']
        print(link)
        datasets.append((name, link))

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

datasets
```

```python
uk_datasets = pd.DataFrame(datasets, columns=['Title', 'Link'])

uk_datasets
```

### List all language names and number of related articles in the order they appear in wikipedia.org: url = 'https://www.wikipedia.org/'

```python
#www-wikipedia-org > div.central-featured

url = 'https://www.wikipedia.org/'
response = requests.get(url)
wiki = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    table = soup.select("body > div.central-featured > div > a")
    #print(table)
    
    for title in table:
        #print(title)
        lang = title.find('strong').text
        num = title.find('bdi').text.replace(u'\xa0', u'')
        #print(lang)
        wiki.append((lang, num))

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

wiki
```

```python
wiki_languages = pd.DataFrame(wiki, columns=['Language', 'Articles'])

wiki_languages
```
